```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles, make_moons

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))

X, Y = make_circles(noise=0.1, factor=0.3, random_state=0)
ax1.scatter(X[:, 0], X[:, 1], c=Y)
ax1.set_title("make_circles")

X, Y = make_moons(noise=0.1, random_state=0)
ax2.scatter(X[:, 0], X[:, 1], c=Y)
ax2.set_title("make_moons")

plt.tight_layout()
plt.show()
```

```python
X, Y = make_circles(n_samples=200, noise=0.1, factor=0.3, random_state=0)
X_test, Y_test = make_circles(n_samples=100, noise=0.1, factor=0.3, random_state=0)
```

```python
model = micrograd.MLP(2, [2, 20, 20, 1])
training_loss, test_loss = [], []
import math

for e in range(20):
    pred = [model(x) for x in X_test]
    loss = [(ypred.log() * y + (-ypred + 1).log() * (1 - y)) * -1 for y, ypred in zip (Y_test, pred)]

    epoch_loss = 0
    for l in loss:
        epoch_loss += l.data
    epoch_loss /= len(loss)
    test_loss.append(epoch_loss)

    pred = [model(x) for x in X]
    loss = [(ypred.log() * y + (-ypred + 1).log() * (1 - y)) * -1 for y, ypred in zip (Y, pred)]

    epoch_loss = 0
    for l in loss:
        l.backward()
        epoch_loss += l.data
    epoch_loss /= len(loss)
    training_loss.append(epoch_loss)

    for p in model.parameters():
        p.data -= 0.001 * p.grad

    print('epoch: {}, training_loss: {}, test_loss: {}'.format(e, training_loss[-1], test_loss[-1]))
```

```python
import matplotlib.pyplot as plt


# Plotting the training loss curve
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(training_loss) + 1), training_loss, marker='o', linestyle='-', color='b')
plt.plot(range(1, len(test_loss) + 1), test_loss, marker='o', linestyle='-', color='r')
plt.title('Training Loss Curve for BCE')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.show()
```

```python
pred = [model(x) for x in X_test]
Y_pred = [1 if p.data > 0.5 else 0 for p in pred]
```

```python
plt.scatter(X_test[:, 0], X_test[:, 1], c=Y_pred)

plt.tight_layout()
plt.show()
```
